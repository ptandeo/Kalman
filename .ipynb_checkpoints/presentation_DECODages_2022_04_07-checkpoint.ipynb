```python
%load_ext autoreload
%autoreload 2

from codes.Kalman import *
from scipy.integrate import odeint
import pandas

%pylab inline
%matplotlib inline
pylab.rcParams['figure.figsize'] = (16, 9)
```

```python
# number of loops
nb_loop = 50
```

```python
# Lotka-Volterra dynamical model
def Lotka_Volterra(x, dx, alpha, beta, gamma, delta):
    dx = zeros((2))
    dx[0] = alpha*x[0]-beta*x[0]*x[1]
    dx[1] = delta*x[0]*x[1]-gamma*x[1]
    return dx

# Lotka-Volterra parameters
x0 = array([1, 1]) # initial condition
dt = 0.1 # integration time step
alpha = 2.0/3; beta = 4.0/3; gamma = 1.0; delta = 1.0 # physical parameters
tab_param = [alpha, beta, gamma, delta]

# time and truth
t   = arange(0.01, nb_loop, dt)
x_t = odeint(Lotka_Volterra, x0, arange(0.01, nb_loop+0.01, dt), args=(alpha, beta, gamma, delta))

# observations
y = x_t.copy()
```

### 1. Dynamical model: schematic representation

```python
plot(t[20:40], x_t[20:40,0], 'b-')
scatter(t[25], x_t[25,0], color='k')
scatter(t[35], x_t[35,0], color='k')
text(t[30], x_t[25,0], '$dt$', verticalalignment='top', size=30)
text(t[35], x_t[30,0], '$dx$', horizontalalignment='left', size=30)
text(t[30], x_t[30,0], '$f$', verticalalignment='top', size=30, color='b')
plot([t[25], t[35]], [x_t[25,0], x_t[25,0]], 'k--')
plot([t[35], t[35]], [x_t[25,0], x_t[35,0]], 'k--')
xticks([t[25], t[35]], ['$t$', '$t+dt$'], size=30)
yticks([x_t[25,0], x_t[35,0]], ['$x(t)$', '$x(t+dt)$'], size=30)
grid()
```

```python
# colors and labels of the components
tab_colors = ['tab:blue', 'tab:orange']
tab_labels = ['$x_1$', '$x_2$', r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$']

# plot the components
plot(t, x_t)
#plot(t, y, 'k.')
xlabel('Time', size=20)
ylabel('Population', size=20)
legend(tab_labels[0:2], loc=1, fontsize='xx-large')
title('Lotka-Volterra', size=20)
grid()
```

### 1. state augmentation (demonstration)

```python
# dynamical model with augmented state
def m(x_past):
    
    tmp = zeros((2,2))
    
    x_future = x_past.copy()
    
    # integration time step
    dt = 0.1
    
    # dynamical model of the state
    tmp = odeint(Lotka_Volterra,
                 x_past[0:2],
                 arange(0, dt+0.01, dt),
                 args=(x_past[2], x_past[3], x_past[4], x_past[5]))
    
    x_future[0:2] = tmp[-1,:]
    
    # dynamical model of the parameters
    x_future[2:] = 0.999 * x_past[2:]
    
    return x_future
```

### 1. state augmentation (real data)

```python
# read data
hare_lynx = array(pandas.read_csv("data/hare_lynx.csv"))
plot(hare_lynx[:,0], hare_lynx[:,1])
plot(hare_lynx[:,0], hare_lynx[:,2])
legend(['Hare', 'Lynx'], loc=1, fontsize='xx-large')
xlabel('Years', size=20)
ylabel('Population (in thousands)', size=20)
grid()
```

```python
# first estimates of the model and the parameters 
from sklearn.linear_model import LinearRegression
X = c_[hare_lynx[0:-1,1], hare_lynx[0:-1,1] * hare_lynx[0:-1,2]]
Y = c_[(hare_lynx[1:,1] - hare_lynx[0:-1,1]) / 0.01]
reg = LinearRegression(fit_intercept=False).fit(X, Y)
alpha = reg.coef_[0,0]
beta = -reg.coef_[0,1]
X = c_[hare_lynx[0:-1,2], hare_lynx[0:-1,1]*hare_lynx[0:-1,2]]
Y = c_[(hare_lynx[1:,2] - hare_lynx[0:-1,2]) / 0.01]
reg = LinearRegression(fit_intercept=False).fit(X, Y)
gamma = -reg.coef_[0,0]
delta = reg.coef_[0,1]
print(alpha, beta, gamma, delta)
```

```python
# dynamical model with augmented state
def m(x_past):
    
    tmp = zeros((2,2))
    
    x_future = x_past.copy()
    
    # integration time step
    dt = 0.08
    
    # dynamical model of the state
    tmp = odeint(Lotka_Volterra,
                 x_past[0:2],
                 arange(0, dt+0.01, dt),
                 args=(x_past[2], x_past[3], x_past[4], x_past[5]))
    
    x_future[0:2] = tmp[-1,:]
    
    # dynamical model of the parameters
    x_future[2:] = 0.999 * x_past[2:]
    
    return x_future
```

```python
x_tmp = zeros([90, 6])
x_tmp[0,:] = c_[hare_lynx[0,1], hare_lynx[0,2], alpha, beta, gamma, delta]
for t in arange(1, 90):
    x_tmp[t] = m(x_tmp[t-1,:])
plot(x_tmp[:,0:2])
grid()
```

```python
# kalman parameters
x0 = c_[hare_lynx[0,1], hare_lynx[0,2], alpha, beta, gamma, delta]
n  = shape(x0)[1]
P0 = eye(n); P0[0,0] = 50; P0[1,1] = 50
H = delete(eye(n), [2,3,4,5], axis=0)
p  = shape(y)[1]
Q = eye(n); Q[0,0] = 50; Q[1,1] = 50
R  = 100*eye(p)
Ne = 1000

# ensemble Kalman filter
x_f, P_f, x_a, P_a, loglik = ensemble_Kalman_filter(hare_lynx[:,1:], x0, P0, m, Q, H, R, Ne)
```
