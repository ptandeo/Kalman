```python
%load_ext autoreload
%autoreload 2

from codes.Kalman import *
from scipy.integrate import odeint
import pandas

%pylab inline
%matplotlib inline
pylab.rcParams['figure.figsize'] = (16, 9)
```

```python
### IMPORTANT PARAMETERS

# variance of the random white noise of z
variance_unobs_comp = 5

# variance of the observation error used in Kalman
variance_obs_comp = 0.0001

# number of SEM iterations
nb_iter_SEM = 1000
```

```python
### READ DATA

# read the datasets
IBTrACS = array(pandas.read_csv('data/test_KATE_2003_stormNb5.csv'))

# time
t = IBTrACS[:,0]*3

# observations
y = IBTrACS.copy()
y = delete(y, 0, axis=1)
```

### V1: $x = [V_{max}, R_{max}, R_{34}, f, z_1]$

```python
# state
z = random.normal(loc=y[:,0]*0, scale=variance_unobs_comp, size=shape(y)[0])
x = c_[y, z]

# shapes
n = shape(x)[1]
p = shape(y)[1]

# colors and labels of the components
tab_colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple']
tab_labels = ['$V_{max}$', '$R_{max}$', '$R_{34}$', '$f$', '$z_1$']

# plot the components
plot(t, x)
xlabel('Hours', size=20)
ylabel('Components of the system', size=20)
legend(tab_labels, loc=1, fontsize='xx-large')
title('Initial x', size=30)
```

```python
# kalman parameters
H = delete(eye(n), 4, axis=0)
R = variance_obs_comp*eye(p)

# stochastic EM
x_s_V1, P_s_V1, M_V1, tab_loglik_V1, x = Kalman_SEM(x, y, H, R, nb_iter_SEM)
```

```python
plot(tab_loglik_V1)
```

```python
# plot the reconstructed components
#for i in range(n):
for i in [0,4]:
    plot(t, x_s_V1[:,i], color=tab_colors[i])
    fill_between(t, x_s_V1[:,i]-1.96*sqrt(P_s_V1[:,i,i]), x_s_V1[:,i]+1.96*sqrt(P_s_V1[:,i,i]),
                 facecolor=tab_colors[i], alpha=0.50)

xlabel('Hours', size=20)
ylabel('Components of the system', size=20)
legend([tab_labels[0], tab_labels[4]], loc=1, fontsize='xx-large')
title('Final x', size=30)
```
