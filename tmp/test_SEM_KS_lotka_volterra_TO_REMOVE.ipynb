```python
%load_ext autoreload
%autoreload 2

from codes.Kalman_filter_smoother import *
from scipy.integrate import odeint
from sklearn.linear_model import LinearRegression

%pylab inline
%matplotlib inline
pylab.rcParams['figure.figsize'] = (16, 9)
```

```python
### IMPORTANT PARAMETERS

# index of the unobserved component
i_unobs_comp = 0

# variance of the random white noise added to the unobserved component in the initial catalog
variance_unobs_comp = 5

# variance of the observations used to generate the catalog and used in the Kalman smoother
variance_obs_comp = 0.0001

# number of Lorenz-63 times
nb_loop = 10

# number of SEM iterations
nb_iter_SEM = 40
```

```python
### GENERATE SIMULATED DATA

# Lotka-Volterra dynamical model
def Lotka_Volterra(x, dx, alpha, beta, gamma, delta):
    dx = zeros((2))
    dx[0] = alpha*x[0]-beta*x[0]*x[1]
    dx[1] = delta*x[0]*x[1]-gamma*S[1]
    return dx

# Lotka-Volterra parameters
x0 = array([1, 1]) # initial condition
dt = 0.1 # integration time step
alpha = 2.0/3; beta = 4.0/3; gamma = 1.0; delta = 1.0 # physical parameters

# time and truth
t   = arange(0.01, nb_loop, dt)
x_t = odeint(Lotka_Volterra, x0, arange(0.01, nb_loop, dt), args=(alpha, beta, gamma, delta))

# observations
y = x_t.copy()
y = delete(y, i_unobs_comp, axis=1)
### ADD NOISE HERE? ###
```

### V1: $x = [z, x_2, x_3]$

```python
# state
z = random.normal(loc=x_t[:,i_unobs_comp]*0, scale=variance_unobs_comp, size=shape(y)[0])
x = c_[z, x_t[:,1], x_t[:,2]]

# shapes
T = len(t)
n = shape(x)[1]
p = shape(y)[1]

# colors and labels of the components
tab_colors = ['tab:blue', 'tab:orange', 'tab:green']
tab_labels = ['$z$', '$x_2$', '$x_3$']

# plot the components
plot(t, x)
xlabel('System loops', size=20)
ylabel('Components of the system', size=20)
legend(tab_labels, loc=1, fontsize='xx-large')
title('Initial x', size=30)
```

```python
### APPLY THE ALGORYTHM

# initialize the log-likelihood
tab_loglik = []

# loop on the SEM iterations
for i in arange(0, nb_iter_SEM):

    print('##### Iteration ' + str(i) + ' #####')
    
    # state of the system
    X = c_[x]
    q = shape(X)[1]
    
    # Kalman parameters
    reg   = LinearRegression(fit_intercept=False).fit(X[:-1,], X[1:,])
    M     = reg.coef_
    Q     = cov((X[1:,] - reg.predict(X[:-1,])).T)
    H     = delete(eye(q), i_unobs_comp, axis=0)
    R     = variance_obs_comp*eye(p)
    
    # Kalman initialization
    if i==0:
        x0 = zeros(q)
        P0 = eye(q)
    else:
        x0 = x_s[0,:]
        P0 = P_s[0,:,:,]
    
    # apply the Kalman smoother
    x_f, P_f, x_a, P_a, x_s, P_s, loglik = Kalman_smoother(t, y, x0, P0, M, Q, H, R)

    # store the log-likelihod
    tab_loglik = append(tab_loglik, sum(loglik))
    
    # simulate the new x
    for k in range(len(x_s)):
        x[k,0:n] = random.multivariate_normal(x_s[k,0:n], P_s[k,0:n,0:n])
```

```python
# check the log-likelihood
plot(tab_loglik)
```

```python
# plot the reconstructed components
plot(t, x)
xlabel('System loops', size=20)
ylabel('Components of the system', size=20)
legend(tab_labels, loc=1, fontsize='xx-large')
title('Final x', size=30)
```
